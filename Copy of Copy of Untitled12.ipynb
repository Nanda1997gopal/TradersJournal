```python
import pandas as pd
de = pd.read_csv('/content/grive/MyDrive/CSV/Book2.csv',encoding="ISO-8859-1")
print(de)
```

```python
de['Age:'].plot(figsize=(16,6))
```

```python
from sklearn import linear_model
import numpy as np
c=de[['Age:']]
d=de['How_many_Sectors_are_there_to_inves_ in_Stock_Market?']
reg=linear_model.LinearRegression()
reg.fit(c,d)
reg.predict(np.array(100).reshape(-1,1))
```

```python
print("value of m",reg.coef_)
print("value of intercept",reg.intercept_)
y=135.78767123*3300+180616.43835616432
print("predicted area is",y)
```

```python
from sklearn.model_selection import train_test_split
c=de[['Age:']]
d=de['How_many_Sectors_are_there_to_inves_ in_Stock_Market?']
X_train , X_test , y_train , y_test=train_test_split(c,d,random_state=0)
print(X_train)
print(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train , y_train)
model.predict(X_test)

```
